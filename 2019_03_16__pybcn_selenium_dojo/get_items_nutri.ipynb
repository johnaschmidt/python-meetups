```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
base_url = 'https://www.urban-nutri-shop.com/'
```

```python
input_pages = [
    {
        'title' : 'Genius Nutrition',
        'product_list_url' : 'https://www.urban-nutri-shop.com/genius-nutrition-complement-alimentaire-musculation-c102x3285230',
        'page_description' : 'Genius, un stage du bio, saveurs naturelles et les meilleurs prix du marché! Qualité Premium assuré. Large choix.  Notre HIT des ventes.\nLa Roll-Royce de la nutrition sportive!'
    },
    {
        'title' : 'HX Premium',
        'product_list_url' : 'https://www.urban-nutri-shop.com/hx-premium-musculation-c102x3667086',
        'page_description' : 'HX Premium, La qualité Premium assurée aux meilleurs prix du marché. Spécialement conçu pour la pratique des sports et de la musculation. Large choix.  Notre HIT des ventes.\nLa Ferrari des compléments alimentaires!'
    },
        {
        'title' : 'Lien Corgenic',
        'product_list_url' : 'https://www.urban-nutri-shop.com/corgenic-musculation-c102x3220383',
        'page_description' : 'Offre: Profite de nos prix bas sur la marque française de qualité Premium!\nProtéines réalisées à partir de matière fraiche. Sub Zero et Whey Supremacy\nPyrocycline TCI brûleur légendaire et bien d\'autres comme V-Type BCAA.'
    }
]
```

```python
output_pages = []

output_pages += [{
    'image_path':'https://static0.tiendeo.com/images/tiendas/399/catalogos/492498/paginas/med3/00001.jpg', 
    'logo': 'https://www.urban-nutri-shop.com/Files/124842/_Oxatis_Design_Img/logo.png', 
    'title': 'COMPLÉMENTS ALIMENTAIRES POUR LE SPORT & LA MUSCULATION', 'start_date':'05/05/2019', 'end_date': '30/06/2019'}]

for page in input_pages:
    source = urlopen(page['product_list_url'])
    content = BeautifulSoup(source,"html.parser")
    
    items = []
    
    items += [{'item_type':'banner','availability': 'InStock', 'image_path':'https://static0.tiendeo.com/images/tiendas/91672/catalogos/492681/paginas/mid2/00001.jpg', 'target_url':'www.tiendeo.com'}]

    for element in content.find_all('td', {'data-pdt-id':True}):
        product = {}
        product['item_type'] = 'product'
        product['image_path'] = None  
        product['thumbnail_path'] = element.find('img', {'class':'imgthumbnail'})
        if product['thumbnail_path']: 
            product['thumbnail_path'] = base_url + product['thumbnail_path'].get('src')
            product['image_path'] = product['thumbnail_path'].replace('09', '03').replace('-small.', '-.')

        product['title'] = element.find('h3', {'class':'PBMainTxt'}).text

        product['detail_url'] = base_url + element.find('a', {'class':'PBLink'}).get('href')

        product['price_sales'] = element.find('span', {'class':'PBSalesPrice'})
        if product['price_sales']: product['price_sales'] = product['price_sales'].text.replace('\xa0', '')

        product['price_before'] = element.find('div', {'class':'PBStrike'})
        if product['price_before']: product['price_before'] = product['price_before'].text.replace('\xa0', '')

        product['discount'] = element.find('span', {'class':'PBDiscount'})
        if product['discount']: product['discount'] = product['discount'].text.replace('\xa0', '')

        product['short_description'] = element.find('span', {'class':'PBShortTxt'})
        if product['short_description']: product['short_description'] = product['short_description'].text

        product['availability'] = element.find('meta', {'itemprop':'availability'})
        if product['availability']: product['availability'] = product['availability'].get('content')

        items += [product]

    page['items'] = items
    
    output_pages += [page]
```

```python
import json

output_json = json.dumps(output_pages, sort_keys=True, separators = (',',':'))

file= open('output_file.json', 'w')

file.write(output_json)

file.close()
```
